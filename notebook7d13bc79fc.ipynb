```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

```python
df = pd.read_csv('/kaggle/input/zomato-bangalore-restaurants/zomato.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df = df.drop(['url','address','phone','menu_item','dish_liked', 'reviews_list'], axis = 1)
df.head()
```

```python
df.info()
```

```python
df.drop_duplicates(inplace = True)
df.shape
```

```python
df['rate'].unique()
```

```python
#removing new, - and / from rate using a function
def handle_rate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
    
df['rate'] = df['rate'].apply(handle_rate)
df['rate'].head()
```

```python
df.rate.isnull().sum()
```

```python
#Filling Null value in Rtae column with mean
df['rate'].fillna(df['rate'].mean(), inplace = True)
df['rate'].isnull().sum()
```

```python
df.info()
```

```python
df.dropna(inplace = True)
df.head()
```

```python
df.rename(columns = {'approx_cost(for two people)': 'Cost_for2','listed_in(type)':'Type','listed_in(city)': 'city'}, inplace = True)
df.head()
```

```python
df['location'].unique()
```

```python
df['city'].unique()
```

```python
df = df.drop(['city'], axis =1)
df.head()
```

```python
df['Cost_for2'].unique()
```

```python
def handle_comma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',','')
        return float(value)
    else:
        return float(value)
    
df['Cost_for2'] = df['Cost_for2'].apply(handle_comma)
df['Cost_for2'].unique()
```

```python
df.head()
```

```python
df['rest_type'].value_counts()
```

```python
rest_types = df['rest_type'].value_counts(ascending = False)
rest_types
```

```python
rest_types_lessth1000 = rest_types[rest_types<1000]
rest_types_lessth1000
```

```python
def handle_rest_type(value):
    if(value in rest_types_lessth1000):
        return 'others'
    else:
        return value
df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
location = df['location'].value_counts()
```

```python
location_lessthn300 = location[location<300]
location_lessthn300
```

```python
def handle_location(value):
    if(value in location_lessthn300):
        return 'others'
    else:
        return value
    
df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
cuisines = df['cuisines'].value_counts(ascending = False)


cuisine_100 = cuisines[cuisines<100]

def handle_cuisines(value):
    if(value in cuisine_100):
        return 'others'
    else:
        return value

df['location'] = df['location'].apply(handle_cuisines)
df['location'].value_counts()
```
